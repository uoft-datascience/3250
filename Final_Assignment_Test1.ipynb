## Import Data

```python
import pandas as pd
from pandas import DataFrame
import pylab as pylab
from pylab import *
import matplotlib.pyplot as pt
import numpy as np
import seaborn as sns
```

```python
card = pd.read_csv("german_credit_card.csv", 
                      parse_dates=False, skip_blank_lines=False, encoding='utf-8', header=0, sep=",")
```

```python
card.head(10)
```

```python
card.columns = [c.replace(' ', '_') for c in card.columns] # replace space with underscore
card.columns = [c.replace('(', '') for c in card.columns]
card.columns = [c.replace(')', '') for c in card.columns]
card.columns = [c.replace('/', '_') for c in card.columns]
card.columns = [c.replace('&', '_') for c in card.columns]
card.head(5)
```

```python
card.describe()
```

```python
pylab.figure()
pylab.hist(card)
pylab.show()
```

```python
card.info()
```

```python
fig, axes = plt.subplots(1, 10, figsize=(22,4))
axes[0].hist(card.Creditability)
axes[0].set_title('Creditability')
axes[1].hist(card.Account_Balance)
axes[1].set_title('Account Balance')
axes[2].hist(card.Duration_of_Credit_month)
axes[2].set_title('Duration_of_Credit_month')
axes[3].hist(card.Payment_Status_of_Previous_Credit)
axes[3].set_title('Payment_Status_of_Previous_Credit')
axes[4].hist(card.Purpose)
axes[4].set_title('Purpose')
axes[5].hist(card.Credit_Amount)
axes[5].set_title('Credit_Amount')
axes[6].hist(card.Value_Savings_Stocks)
axes[6].set_title('Value_Savings_Stocks')
axes[7].hist(card.Length_of_current_employment)
axes[7].set_title('Length_of_current_employment')
axes[8].hist(card.Instalment_per_cent)
axes[8].set_title('Instalment_per_cent')
axes[9].hist(card.Sex___Marital_Status)
axes[9].set_title('Sex___Marital_Status')
plt.show()
```
